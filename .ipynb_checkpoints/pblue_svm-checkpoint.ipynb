```python
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.preprocessing import LabelEncoder
from collections import defaultdict
from nltk.corpus import wordnet as wn
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import model_selection, naive_bayes, svm
from sklearn.metrics import accuracy_score
import regex as re
from collections import Counter
```

```python
np.random.seed(500)
```

## 1. Connect to the DB and export the items table to a Pandas DF.

```python
import pymysql.cursors
from pandas import DataFrame

connection = pymysql.connect(host='127.0.0.1',
                             user='root',
                             password='john1snico',
                             database='pblue',
                             cursorclass=pymysql.cursors.DictCursor)

query = "SELECT r.id, r.item_id, r.recommended, r.pros, r.cons, r.comments, i.name FROM pblue.reviews r join pblue.items i on r.item_id = i.id;"

cursor = connection.cursor()
cursor.execute(query)
df = DataFrame(cursor.fetchall())

def combined_features(row):
    return row['name'] + ' ' + row['pros'] + ' ' + row['cons'] + ' ' + row['comments']

df['all_features'] = df.apply(combined_features, axis=1)

cols = df[['recommended', 'all_features']]
corpus = cols.copy()
```

```python
corpus.head()
```

```python
# Remove blank rows if any.
corpus['all_features'].dropna(inplace=True)

# Change all the text to lower case. This is required as python interprets 'dog' and 'DOG' differently
corpus['all_features'] = [entry.lower() for entry in corpus['all_features']]

# Tokenization : In this each entry in the corpus will be broken into set of words
corpus['all_features']= [word_tokenize(entry) for entry in corpus['all_features']]

# Remove Stop words, Non-Numeric and perfom Word Stemming/Lemmenting.
# WordNetLemmatizer requires Pos tags to understand if the word is noun or verb or adjective etc. By default it is set to Noun
tag_map = defaultdict(lambda : wn.NOUN)
tag_map['J'] = wn.ADJ
tag_map['V'] = wn.VERB
tag_map['R'] = wn.ADV

for index,entry in enumerate(corpus['all_features']):
    final_words = []
    
    # Initialize WordNetLemmatizer()
    word_lemmatized = WordNetLemmatizer()
    
    # pos_tag function below will provide the 'tag' i.e if the word is Noun(N) or Verb(V) or something else.
    for word, tag in pos_tag(entry):
        
        if word not in stopwords.words('english') and word.isalpha():
            word_final = word_lemmatized.lemmatize(word,tag_map[tag[0]])
            final_words.append(word_final)
            
    corpus.loc[index,'text_final'] = str(final_words)
```

```python
# Prep Train and Test Data sets
train_X, test_X, train_Y, test_Y = model_selection.train_test_split(corpus['text_final'],corpus['recommended'],test_size=0.3)
```

```python
# Turn string into numeric
encoder = LabelEncoder()
train_Y = encoder.fit_transform(train_Y)
test_Y = encoder.fit_transform(test_Y)
```

```python
# Vectorize the reviews
tfidf_vect = TfidfVectorizer(max_features=5000)
tfidf_vect.fit(corpus['text_final'])

train_X_tfidf = tfidf_vect.transform(train_X)
test_X_tfidf = tfidf_vect.transform(test_X)
```

```python
print(tfidf_vect.vocabulary_)
```

```python
print(train_X_tfidf)
```

## 2. Build a classifier using an SVM to identify whether a comment results to a recommendation or not

```python
# Classifier - Algorithm - SVM
# fit the training dataset on the classifier
SVM = svm.SVC(C=1.0, kernel='linear', degree=3, gamma='auto')
SVM.fit(train_X_tfidf,train_Y)

# predict the labels on validation dataset
predictions_SVM = SVM.predict(test_X_tfidf)

# Use accuracy_score function to get the accuracy
print("SVM Accuracy Score -> ",accuracy_score(predictions_SVM, test_Y)*100)
```

## 4.Identify the top words for the pros and cons columns

```python
import pymysql.cursors
from pandas import DataFrame

connection = pymysql.connect(host='127.0.0.1',
                             user='root',
                             password='john1snico',
                             database='pblue',
                             cursorclass=pymysql.cursors.DictCursor)

query = "SELECT r.pros FROM pblue.reviews r;"

cursor = connection.cursor()
cursor.execute(query)
word_df = DataFrame(cursor.fetchall())
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')

newStopWords = ["class", 
                  "know", 
                  "make", 
                  "give", 
                  "students", 
                  "gives", 
                  "student", 
                  "lot", 
                  "really",
                  "lecture",
                  "well",
                  "knows",
                  ]
stop.extend(newStopWords)


word_df['pros'] = word_df['pros'].str.lower()
word_df['pros'] = word_df['pros'].str.replace(',', '')
word_df['pros'] = word_df['pros'].str.replace('.', '')
word_df['pros'] = word_df['pros'].str.strip()


word_df['pros_c'] = word_df['pros'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))

word_df['words'] = [len(x) for x in word_df['pros_c'].tolist()]
```

```python
words = ''
for i in word_df.pros_c.values:
    words += '{} '.format(i.lower())
    
wd = pd.DataFrame(Counter(words.split()).most_common(200), columns=['word', 'frequency'])

wd
```
