### Explain what KMeans inertia is and how different it is from DBSCAN epsilon.

#### K-means Inertia vs. DBSCAN Epsilon

K-means Inertia

- Inertia is the sum of squared error for each cluster. It is the distance of cluster points towards its centroid
- The results of a K-means cluster are dependent on where you make your "random selection". The closer the data rows are to one another the more they will be clustered.

DBSCAN Epsilon

- Represents the maximum distance that points can be from each other to be considered a
  cluster

```python
import numpy as np

X = np.array(df[["lat", "lng"]].values.tolist())
plt.scatter(X[:, 0], X[:, 1], s=50)
```

```python
from sklearn.cluster import KMeans


sum_of_squared_distances = []
K = range(1,30)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(X)
    sum_of_squared_distances.append(km.inertia_)

plt.plot(K, sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
kmeans = KMeans(n_clusters=6)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

```python
import seaborn as sns

df["kmeans_cluster"] = y_kmeans
palette = sns.color_palette("pastel", 6).as_hex()
color_getter = lambda x: palette[int(x)]
df["k_color"] = df["kmeans_cluster"].map(color_getter)
df.head()
```

```python
basemap = folium.Map(location=[14.5562297,121.0204097], tiles="cartodbdark_matter", zoom_start=15)

for index, row in df.iterrows():
    folium.CircleMarker([row["lat"], row["lng"]], radius=5,
    fill=True,
    color =row["k_color"],
                        popup = folium.Popup(row["name"], parse_html=True),
    fill_color=row["k_color"]).add_to(basemap)
    
basemap.save('fast_food_k.html')

basemap
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.005, min_samples=5).fit(X)
y_dbscan = dbscan.labels_

num_clusters = len(set(y_dbscan))
print(f'Number of clusters : {num_clusters}')
```

```python
y_dbscan
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, s=50, cmap='viridis')
```

```python
df["dbscan_cluster"] = y_dbscan
palette = sns.color_palette("pastel", len(y_dbscan)).as_hex()
color_getter = lambda x: palette[int(x)]
df["db_color"] = df["dbscan_cluster"].map(color_getter)
df.head()
```

```python
df = df[df["dbscan_cluster"] > -1]
```
